```python
# time showing
import time
import IPython as pt
current_time = time.asctime(time.localtime())[11:20]
year = time.asctime(time.localtime())[20:]
date = time.asctime(time.localtime()) [:10]+' '+year
print(f'date: {date}')
while True:
    pt.display.clear_output(wait=True)
    current_time = time.asctime(time.localtime())[11:20]
    time.sleep(1)
    print(f'time: {current_time} ')
```
