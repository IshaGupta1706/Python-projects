```python
author = []
book_name = []
price = []
rating = []
num_cust = []

for i,j in enumerate(soup.find_all('div',{'class':'_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y'})):
    if i%2==0:
        book_name.append(j.text)
    else:
        author.append(j.text)
        
        
for i,j in enumerate(soup.find_all('span',{'class':'_cDEzb_p13n-sc-price_3mJ9Z'})):
    price.append(j.text.replace('₹',''))

    
    
for i,j in enumerate(soup.find_all('div',{'class':'a-icon-row'})):
    rating.append((float(j.text[:3])))   
    
    
for i,j in enumerate(soup.find_all('div',{'class':'a-icon-row'})):
    num_cust.append(int(float(j.text[19:].replace(',',''))))
```

```python
def amazon(page_no):
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
    url2 =f'''https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_{page_no}_books?ie=UTF8&pg={page_no}'''
    r = requests.get(url2,headers=headers) 
    data = r.text    
    soup = BeautifulSoup(data,'html.parser')
    author = []
    book_name = []
    price = []
    rating = []
    num_cust = []
    for i,j in enumerate(soup.find_all('div',{'class':'_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y'})):
        if i%2==0:
            book_name.append(j.text)
        else:
            author.append(j.text)


    for i,j in enumerate(soup.find_all('span',{'class':'_cDEzb_p13n-sc-price_3mJ9Z'})):
        price.append(j.text.replace('₹',''))



    for i,j in enumerate(soup.find_all('div',{'class':'a-icon-row'})):
        rating.append((float(j.text[:3])))   


    for i,j in enumerate(soup.find_all('div',{'class':'a-icon-row'})):
        num_cust.append(int(float(j.text[19:].replace(',',''))))
    final_data = {'Author':author,
                     'Book title': book_name,
                     'Price': price,
                     'Rating':rating,
                     'Total_customer_review': num_cust}
#     print(len(author))
#     print(len(book_name))
#     print(len(price))
#     print(len(rating))
#     print(len(num_cust))
    

    book_df = pd.DataFrame(final_data)    
    return book_df   
```
